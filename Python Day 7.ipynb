```python
# the string is palindrome or not ?
def palindrome(n):
    if n==n[::-1]:
        return "palindrome"
    else:
        return "not palindrome"
palindrome("helllo")
```

```python
def pat(n):
    for i in range(n):
        for j in range(2**i):
            print("a",end=" ")
        for j in range(i+1):
            print("b",end=" ")
        print()
pat(5)
```

```python
# factorial of 5
# 5*4*3*2*1
# ans = 120
# NO argument with return
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f
fact(5)
    
        
```

```python
# sum of natural numbers
# n*(n+1)/2
# 10
# 10*11/2
# 55
def sum(n):
    f=1
    for i in range(1,n+1):
        f=n*(n+1)/2
    return f
sum(10)

```

```python
def sum(n):
    f=0
    for i in range(1,n+1):
        f=f+i
    return f
sum(15)
```
