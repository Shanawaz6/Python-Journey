# Game 



```python
import random
lucky_number=random.randint(5,10)
user_num=int(input("enter the number"))
c=1
while user_num!=lucky_number:
    if user_num<lucky_number:
         print("Choose Higher")
    else:
        print("choose lower")
    user_num=int(input())
    c+=1
else:
    print("you won the game")
    print("yout attempt",c)

```

## Continue

```python
#continue will remove the that number and continue other no.s
for i in range(2,10):
    if i==7:
        continue
    print(i)
```

# Nested Loop

```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

```python
for i in range(1,10):
    for j in range(1,10):
        print(i,j)
```

```python
n=int(input())
for i in range(1,n):  #outer loop
    for j in range(1,i+1):  #inner loop
        print('*',end='')
    print()
```
